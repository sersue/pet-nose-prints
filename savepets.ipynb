```
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread('/gdrive/My Drive/001.png',cv2.IMREAD_GRAYSCALE) #흑백화
kernel = np.array([[0,-1,0],[-1,5,-1],[0,-1,0]])

image_sharp = cv2.filter2D(image,-1,kernel) # 선명하게

plt.imshow(image_sharp,cmap="gray"),plt.axis("off")
plt.show()
```

```
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('/gdrive/My Drive/001.png',0);

# contrast limit가 2이고 title의 size는 8X8
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img2 = clahe.apply(img)

img = cv2.resize(img,(400,400))
img2 = cv2.resize(img2,(400,400))

dst = np.hstack((img, img2))
cv2.imshow('img',dst)
cv2.waitKey()
cv2.destroyAllWindows()
```
